```python
import scipy.stats as stats
```

```python
d=np.random.normal(10,30,90)
e=np.random.normal(15,46,40)
d
```

Perform sample t test

```python
e
```

```python
t_statistic,p_value=stats.ttest_ind(d,e)
print(t_statistic)
print(p_value)
```

Probability Distributions with SciPy

```python
d1=np.random.normal(16,30,40)
e1=np.random.normal(20,60,40)
```

```python
import matplotlib.pyplot as plt
plt.hist(d1)
```

```python
plt.hist(e1)
```

```python
sum=np.sum(d1)
sum
```

```python
product=np.product(d1)
product
```

```python
root=np.sqrt(d1)
root
```

```python
from scipy import stats
```

```python
s,p=stats.pearsonr(d1,e1)
```

```python
print(s,p)
```

```python
maximum=max(d1)
```

```python
maximum
```

```python
minimum=min(e1)
minimum
```

#Use scipy library to calculate Skewness, Kurtosis

```python
skewness=stats.skew(d1)
skewness
```

```python
Kurtosis=stats.kurtosis(d1)
Kurtosis
```

```python
print(minimum)
print(maximum)
print(sum)
print(product)
print(root)
print(skewness)
print(Kurtosis)
```
